```python
import cv2
import torch

# YOLOv5 모델을 GitHub에서 직접 불러오기
# GitHub에서 YOLOv5 모델 불러오기
model = torch.hub.load('ultralytics/yolov5', 'custom', path='C:/Users/smhrd/Desktop/project_OBD/yolov5/runs/train/exp10/weights/best.pt', source='github')


# 웹캠 캡처 객체 생성 (0은 기본 카메라)
cap = cv2.VideoCapture(0)

# 웹캠이 정상적으로 열렸는지 확인
if not cap.isOpened():
    print("웹캠을 열 수 없습니다.")
    exit()

while True:
    # 프레임 읽기
    ret, frame = cap.read()
    if not ret:
        print("프레임을 가져올 수 없습니다.")
        break

    # YOLOv5 모델로 추론 수행 (이미지 크기 조정 가능)
    results = model(frame)

    # 추론 결과에서 바운딩 박스와 레이블 표시
    for *box, conf, cls in results.xyxy[0]:  # xyxy 형식의 바운딩 박스 좌표
        # 바운딩 박스 좌표 추출
        x1, y1, x2, y2 = map(int, box)
        label = f'{model.names[int(cls)]} {conf:.2f}'  # 클래스 이름과 신뢰도

        # 바운딩 박스 그리기
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # 레이블 텍스트 추가
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2)

    # 결과 프레임을 화면에 표시
    cv2.imshow('YOLOv5 Webcam', frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 웹캠과 창 닫기
cap.release()
cv2.destroyAllWindows()
```
