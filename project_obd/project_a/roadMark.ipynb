```python
import cv2

# 사용 가능한 카메라 장치 확인 함수
def list_available_cameras():
    available_cameras = []
    for i in range(5):  # 최대 5개의 카메라 장치 확인
        cap = cv2.VideoCapture(i)
    
        if cap.isOpened():
            available_cameras.append(i)
            cap.release()
    return available_cameras

# 사용 가능한 카메라 장치 출력
cameras = list_available_cameras()
print("Available Camera Devices:", cameras)
```

```python
import torch

# ONNX 모델 경로 설정
model_path = "yolov5s.onnx"  # ONNX 모델 경로

# YOLOv5 모델 로드 (ONNX 기반)
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)

# 웹캠 입력 설정 (source=0은 기본 웹캠을 의미)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()  # 웹캠에서 프레임 읽기
    if not ret:
        print("웹캠에서 영상을 읽을 수 없습니다.")
        break

    # YOLOv5 모델로 예측 수행
    results = model(frame)

    # 결과 시각화 및 출력
    results.render()  # 결과를 이미지에 그리기
    
    # results.ims로 변경하여 이미지 출력
    for img in results.ims:
        cv2.imshow('YOLOv5 Detection', img)  # 예측 결과를 화면에 표시

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import pathlib
import torch
import cv2

# PosixPath를 WindowsPath로 대체
temp = pathlib.PosixPath
pathlib.PosixPath = pathlib.WindowsPath

# YOLOv5 모델 로드
model = torch.hub.load('ultralytics/yolov5', 'custom', path='C:/Users/smhrd/kimsj020406/yolov5/runs/train/exp14/weights/best.pt')
model.conf = 0.25  # 신뢰도 임계값 설정

# 웹캠 열기 (0번 카메라)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("웹캠을 열 수 없습니다.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("프레임을 읽을 수 없습니다.")
        break

    # YOLOv5 모델로 추론 수행
    results = model(frame)

    # 결과 데이터 가져오기 (바운딩 박스, 클래스, 신뢰도)
    detections = results.xyxy[0].cpu().numpy()  # [x1, y1, x2, y2, confidence, class]

    for detection in detections:
        x1, y1, x2, y2, conf, cls = detection
        class_name = model.names[int(cls)]  # 클래스 이름

        # 바운딩 박스 그리기
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        
        # 클래스 이름 및 신뢰도 표시
        label = f"{class_name} {conf:.2f}"
        cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # 결과 화면 출력
    cv2.imshow('YOLOv5 Real-Time Detection', frame)

     # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 리소스 해제
cap.release()
cv2.destroyAllWindows()
```
